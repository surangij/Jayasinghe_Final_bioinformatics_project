```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord

def read_csv(file_path):
    M5C_positions = []
    with open(file_path, 'r') as csv_file:
        next(csv_file)
        for line in csv_file:
            row = line.strip().split(",")
            position = int(row[0])
            M5C_positions.append(position)
    print("These are my positions:", M5C_positions)
    return M5C_positions

def parse_file(fasta_file, M5C_positions, output_filename):
    context_records = []
    with open(fasta_file, "r") as file:
        genome = next(SeqIO.parse(file, "fasta"))
        for position in M5C_positions:
            context = genome.seq[(position - 20):(position + 20)]
            record_id = "Context_position_" + str(position)
            description = "Sequence context centered at position " + str(position)
            context_record = SeqRecord(context, id=record_id, description=description)
            context_records.append(context_record)
    with open(output_filename, "w") as output_file:
        SeqIO.write(context_records, output_file, "fasta")
    print("Multifasta file with sequences has been created.")

fasta_file = "Heliobacter_pylori_J99_ASM98269v1.fasta"
output_filename = "output.fasta"
csv_file_path = "SRR15720537_primrose_confident_sites.csv"

# Read CSV file to get positions
M5C_positions = read_csv(csv_file_path)

# Parse fasta file and create output
parse_file(fasta_file, M5C_positions, output_filename)
```
